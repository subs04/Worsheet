```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("/content/sample_data/Titanic-Dataset.csv")
df.head()
```

```python
fare = df[['Fare']]
fare.head()

```

```python
class_age = df[['Pclass', 'Age']]
class_age.head()
```

```python
survived_gender = df[['Survived', 'Sex']]
survived_gender.head()
```

```python
# 1) Rows where Fare > 100
fare_gt_100 = df[df['Fare'] > 100]
fare_gt_100.head()
```

```python
#2 Rows where Pclass = 1
first_class = df[df['Pclass'] == 1]
first_class.head()
```

```python
female_under_18 = df[(df['Sex'] == 'female') & (df['Age'] < 18)]
female_under_18.head()
```

```python
embarked_c_or_s = df[df['Embarked'].isin(['C', 'S'])]
embarked_c_or_s.head()
```

```python
first_second_class = df[df['Pclass'].isin([1, 2])]
first_second_class.head()
```

```python
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)
print(df["Age"])
```

```python
df['fare_per_year'] = df['Fare'] / df['Age']
df.head()
```

```python
high_fare_age = df[df['fare_per_year'] > 5]
high_fare_age.head()
```

```python
high_fare_age_sorted = high_fare_age.sort_values(by='fare_per_year', ascending=False)
high_fare_age_sorted[['Name', 'fare_per_year']].head()
```

```python
#Adult males (age 18) who paid highest fare/class
# 1. Add fare_per_class = Fare / Pclass
df['fare_per_class'] = df['Fare'] / df['Pclass']
print(df['fare_per_class'] .head())
```

```python
#2. Take adult males only
adult_males = df[(df['Sex'] == 'male') & (df['Age'] >= 18)]

print("\n=== Adult male passengers (Age >= 18) ===")
print(adult_males.head())
print("\nTotal adult male passengers:", len(adult_males))
```

```python
#1 Total Fare
total_fare = df['Fare'].sum()
print("\nTotal Fare of all passengers:", total_fare)
```

```python
fare_by_class = df.groupby('Pclass')['Fare'].sum()

print("\n=== Total Fare by Class ===")
print(fare_by_class)
```

```python
percentage_fare = (fare_by_class / total_fare) * 100

print("\n=== Percentage of Total Fare by Class ===")
print(percentage_fare)
```

```python
def age_group(x):
    if x < 18:
        return 'child'
    elif x < 65:
        return 'adult'
    else:
        return 'senior'

df['age_group'] = df['Age'].apply(age_group)
age_counts = df['age_group'].value_counts()
print("\n=== Age Group Counts ===")
print(age_counts)
```

```python
age_percent = (age_counts / len(df)) * 100

print("\n=== Age Group Percentages ===")
print(age_percent)
```
