```python
#Cell 5: To-Do-4 — Cost Function (MSE)
def cost_function(X, Y, W):
    """
    Computes Mean Squared Error Cost
    """
    m = len(Y)
    Y_pred = np.dot(X, W)
    cost = (1 / (2 * m)) * np.sum((Y_pred - Y) ** 2)
    return cost
```

```python
#Cell 6: To-Do-5 — Cost Function Test Case
X_test_case = np.array([[1, 2], [3, 4], [5, 6]])
Y_test_case = np.array([3, 7, 11])
W_test_case = np.array([1, 1])

cost = cost_function(X_test_case, Y_test_case, W_test_case)

if cost == 0:
    print("Proceed Further")
else:
    print("Something went wrong")

print("Cost function output:", cost)
```

```python
#Cell 7: To-Do-6 — Gradient Descent Implementation
def gradient_descent(X, Y, W, alpha, iterations):
    """
    Gradient Descent Algorithm
    """
    m = len(Y)
    cost_history = []

    for i in range(iterations):
        # Step 1: Prediction
        Y_pred = np.dot(X, W)

        # Step 2: Loss
        loss = Y_pred - Y

        # Step 3: Gradient
        dw = (1 / m) * np.dot(X.T, loss)

        # Step 4: Update weights
        W = W - alpha * dw

        # Step 5: Store cost
        cost = cost_function(X, Y, W)
        cost_history.append(cost)

    return W, cost_history
```

```python
#Cell 8: To-Do-7 — Gradient Descent Test
np.random.seed(0)

X_random = np.random.rand(100, 3)
Y_random = np.random.rand(100)
W_random = np.random.rand(3)

alpha = 0.01
iterations = 1000

final_params, cost_history = gradient_descent(
    X_random, Y_random, W_random, alpha, iterations
)

print("Final Parameters:", final_params)
print("Final Cost:", cost_history[-1])
```

```python
#Cell 9: To-Do-8 — RMSE Implementation
def rmse(Y, Y_pred):
    error = np.sqrt(np.mean((Y - Y_pred) ** 2))
    return error
```

```python
#Cell 10: To-Do-9 — R² Score Implementation
def r2(Y, Y_pred):
    mean_y = np.mean(Y)
    ss_tot = np.sum((Y - mean_y) ** 2)
    ss_res = np.sum((Y - Y_pred) ** 2)
    r2_score = 1 - (ss_res / ss_tot)
    return r2_score
```

```python
#Cell 11: To-Do-10 — Main Function
def main():
    # Load data
    data = pd.read_csv("/content/sample_data/student (1).csv")

    # Features and target
    X = data[['Math', 'Reading']].values
    Y = data['Writing'].values

    # Train-test split
    X_train, X_test, Y_train, Y_test = train_test_split(
        X, Y, test_size=0.2, random_state=42
    )

    # Initialize weights and hyperparameters
    W = np.zeros(X_train.shape[1])
    alpha = 0.00001
    iterations = 1000

    # Train model
    W_optimal, cost_history = gradient_descent(
        X_train, Y_train, W, alpha, iterations
    )

    # Predictions
    Y_pred = np.dot(X_test, W_optimal)

    # Evaluation
    model_rmse = rmse(Y_test, Y_pred)
    model_r2 = r2(Y_test, Y_pred)

    # Results
    print("Final Weights:", W_optimal)
    print("Cost History (First 10):", cost_history[:10])
    print("RMSE:", model_rmse)
    print("R2 Score:", model_r2)

if __name__ == "__main__":
    main()
```
