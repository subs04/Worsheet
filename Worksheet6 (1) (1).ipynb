```python
"""
2. Task To Do:
• Implement the Log - loss Function by completing the code or writing your own function.
• Make sure you pass the test case.
"""
def log_loss(y_true, y_pred):
    """
    Computes log loss for true target value and predicted value
    """
    import numpy as np

    # Ensure y_pred is clipped to avoid log(0)
    y_pred = np.clip(y_pred, 1e-10, 1 - 1e-10)

    loss = -(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
    return loss
```

```python
"""
3. Task To Do:
• Implement the Cost Function by completing the code or writing your own function.
• Make sure you pass the test case.
"""
def cost_function(y_true, y_pred):
    """
    Computes log loss for inputs true value and predicted value
    """
    assert len(y_true) == len(y_pred)

    n = len(y_true)
    loss_vec = log_loss(y_true, y_pred)
    cost = np.sum(loss_vec) / n

    return cost
```

```python
"""
4. Task To Do:
• Implement the vectorized cost function costfunction logreg Function by completing the
code or writing your own function.
• Make sure you pass the test case.
"""
def costfunction_logreg(X, y, w, b):
    """
    Computes the cost function for logistic regression
    """
    z = np.dot(X, w) + b
    y_pred = logistic_function(z)

    cost = cost_function(y, y_pred)
    return cost
```

```python
"""
5. Task To Do:
• Implement the compute gradient Function by completing the code or writing your own
function.
• Make sure you pass the test case.
"""
def compute_gradient(X, y, w, b):
    """
    Computes gradients of weights and bias
    """
    n = X.shape[0]

    y_pred = logistic_function(np.dot(X, w) + b)   # Your Code Here

    grad_w = (1/n) * np.dot(X.T, (y_pred - y))     # Your Code Here
    grad_b = (1/n) * np.sum(y_pred - y)            # Your Code Here

    return grad_w, grad_b
```

```python
"""
6. Task To Do:
• Implement the gradient descent Function by completing the code or writing your own
function.
• Make sure you pass the test case.
"""
def gradient_descent(X, y, w, b, alpha, n_iter, show_cost=False, show_params=True):
    """
    Implements batch gradient descent for logistic regression
    """
    cost_history = []
    params_history = []

    for i in range(n_iter):

        # Compute gradients
        grad_w, grad_b = compute_gradient(X, y, w, b)

        # Update weights and bias
        w -= alpha * grad_w
        b -= alpha * grad_b

        # Compute cost
        cost = costfunction_logreg(X, y, w, b)

        cost_history.append(cost)
        params_history.append((w.copy(), b))

    return w, b, cost_history, params_history
```

```python
"""
7. Task To Do:
• Implement the prediction Function by completing the code or writing your own function.
• Make sure you pass the test case.
"""
def prediction(X, w, b, threshold=0.5):

    #Predicts binary class labels

    z = np.dot(X, w) + b
    y_test_prob = logistic_function(z)

    y_pred = np.where(y_test_prob >= threshold, 1, 0)
    return y_pred
```

```python
"""
8. Task To Do:
• Implement evaluate classification Function by completing the code or writing your own
function.
• Make sure you pass the test case.
"""
def evaluate_classification(y_true, y_pred):

    #Computes confusion matrix, precision, recall, and F1-score


    TP = np.sum((y_true == 1) & (y_pred == 1))
    TN = np.sum((y_true == 0) & (y_pred == 0))   # Your Code Here
    FP = np.sum((y_true == 0) & (y_pred == 1))   # Your Code Here
    FN = np.sum((y_true == 1) & (y_pred == 0))   # Your Code Here

    confusion_matrix = np.array([[TN, FP],
                                 [FN, TP]])

    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0        # Your Code Here
    f1_score = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0  # Your Code Here

    return confusion_matrix, precision, recall, f1_score
```
