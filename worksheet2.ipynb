```python
import pandas as pd

df = pd.read_csv("/content/Titanic-Dataset.csv")

subset = df[['Name','Pclass','Sex','Age','Fare','Survived']]
subset_first_class = subset[subset['Pclass'] == 1]

mean_fare = subset_first_class['Fare'].mean()
median_fare = subset_first_class['Fare'].median()
max_fare = subset_first_class['Fare'].max()
min_fare = subset_first_class['Fare'].min()

mean_fare, median_fare, max_fare, min_fare
```

```python
subset_first_class['Age'].isnull().sum()

clean_age_df = subset_first_class.dropna(subset=['Age'])
clean_age_df
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/Titanic-Dataset.csv")

survival_by_gender = df.groupby('Sex')['Survived'].mean()
print(survival_by_gender)

sns.barplot(x=df['Sex'], y=df['Survived'])
plt.title("Survival Rate by Gender")
plt.show()
```

```python
sns.barplot(x=df['Embarked'], y=df['Survived'])
plt.title("Survival Rate by Port of Embarkation")
plt.show()

```
